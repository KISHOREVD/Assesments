###### Department wise codekata performence (pie chart)

```python
for_pie=df.groupby("Department")["CodeKata Score"].mean().reset_index()
for_pie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,6))
plt.pie(for_pie["CodeKata Score"],labels=for_pie["Department"],autopct="%1.2f%%",colors=['g','c','r'])
plt.show()
```
